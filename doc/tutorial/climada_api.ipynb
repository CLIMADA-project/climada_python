# Data API

So far, there are two types of queries implemented, one for TropCyclones and one for BlackMarble.\
Both are following the respective tutorial notebooks, BlackMarble very closely the example of Turkey and Iran, only it's restricted to one country at a time.\
TropCyclone is looking for Tracks from the given provider (in the given basin) that have been active in the given country's area (country border shape plus margin in degree lat/lon).
For api call details, see [here](https:/climada.ethz.ch/rest/docs).

```python
from climada.util.api_client import DataApiClient
client = DataApiClient()
```

```python
bm_turks = client.get_black_marble(country='Turks and Caicos Islands', resolution=0.3)
bm_turks.check()
```

```python
bm_turks.plot_hexbin()
```

```python
tc_turks = client.get_trop_cyclone(country='Turks and Caicos Islands', gridsize=100, margin=0.1)
tc_turks.check()
```

```python
tc_turks.plot_intensity(0)
```
